```python
# Import spacy
import spacy

# Install English language model
!spacy download fr_dep_news_trf

# Import os to upload documents and metadata
import os

# Load spaCy visualizer
from spacy import displacy

# Import pandas DataFrame packages
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

# Import graphing package
import plotly.graph_objects as go
import plotly.express as px
```

```python
os.getcwd()
```

```python
texts = []
file_names = []

directory_path = '/Users/mathildecontreras/Documents/CD_A4_Pasteur/Corpus_Pasteur'
for _file_name in os.listdir(directory_path):
    if _file_name.endswith('.txt'):
        file_path = os.path.join(directory_path, _file_name)
        with open(file_path, 'r', encoding='utf-8') as file:
            texts.append(file.read())
            file_names.append(_file_name)
```

```python
d = {'Filename':file_names,'Text':texts}
letter_df = pd.DataFrame(d)
letter_df.head(12)
```

```python
letter_df['Text'] = letter_df['Text'].str.replace('\s+', ' ', regex=True).str.strip()
letter_df.head(12)
```

```python
metadata_df = pd.read_csv('Metadata_pasteur.csv')
metadata_df.head()
```

```python
letter_df['Filename'] = letter_df['Filename'].str.replace('.txt', '', regex=True)
```

```python
metadata_df.rename(columns={"LetterID": "Filename"}, inplace=True)	
```

```python
metadata_df.head(12)
```

```python
final_letter_df = metadata_df.merge(letter_df,on='Filename')
```

```python
final_letter_df.head()
```

```python
nlp = spacy.load('fr_core_news_sm')
```

```python
print(nlp.pipe_names)
```

```python
phrase = "Ceci est une phrase exemple"

# Call the nlp model on the sentence
doc = nlp(phrase)
# Loop through each token in doc object
for token in doc:
    # Print text and part of speech for each
    print(token.text, token.pos_)
```

```python
def process_text(text):
    return nlp(text)
```

```python
final_letter_df['Doc'] = final_letter_df['Text'].apply(process_text)
```

```python
# Define a function to retrieve tokens from a doc object
def get_token(doc):
    return [(token.text) for token in doc]
```

```python
# Run the token retrieval function on the doc objects in the dataframe
final_letter_df['Tokens'] = final_letter_df['Doc'].apply(get_token)
final_letter_df.head()
```

```python
tokens = final_letter_df[['Text', 'Tokens']].copy()
tokens.head()
```

```python
# Define a function to retrieve lemmas from a doc object
def get_lemma(doc):
    return [(token.lemma_) for token in doc]

# Run the lemma retrieval function on the doc objects in the dataframe
final_letter_df['Lemmas'] = final_letter_df['Doc'].apply(get_lemma)
```

```python
print(f'"Parens" appears in the text tokens column ' + str(final_letter_df['Tokens'].apply(lambda x: x.count('parens')).sum()) + ' times.')
print(f'"Parens" appears in the lemmas column ' + str(final_letter_df['Lemmas'].apply(lambda x: x.count('parens')).sum()) + ' times.')
```

```python
# Define a function to retrieve lemmas from a doc object
def get_pos(doc):
    #Return the coarse- and fine-grained part of speech text for each token in the doc
    return [(token.pos_, token.tag_) for token in doc]

# Define a function to retrieve parts of speech from a doc object
final_letter_df['POS'] = final_letter_df['Doc'].apply(get_pos)
```

```python
list(final_letter_df['POS'])
```

```python
spacy.explain("IN")
```

```python
# Define function to extract proper nouns from Doc object
def extract_proper_nouns(doc):
    return [token.text for token in doc if token.pos_ == 'PROPN']

# Apply function to Doc column and store resulting proper nouns in new column
final_letter_df['Proper_Nouns'] = final_letter_df['Doc'].apply(extract_proper_nouns)
```

```python
list(final_letter_df.loc[[2, 2], 'Proper_Nouns'])
```

```python
# Get all NE labels and assign to variable
labels = nlp.get_pipe("ner").labels

# Print each label and its description
for label in labels:
    print(label + ' : ' + spacy.explain(label))
```

```python
# Define function to extract named entities from doc objects
def extract_named_entities(doc):
    return [ent.label_ for ent in doc.ents]

# Apply function to Doc column and store resulting named entities in new column
final_letter_df['Named_Entities'] = final_letter_df['Doc'].apply(extract_named_entities)
final_letter_df['Named_Entities']
```

```python
# Extract the first Doc object
doc = final_letter_df['Doc'][0]

# Visualize named entity tagging in a single paper
displacy.render(doc, style='ent', jupyter=True)
```
