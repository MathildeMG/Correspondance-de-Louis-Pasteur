```python
# Import spacy
import spacy

# Install English language model
!spacy download fr_core_news_sm

# Import os to upload documents and metadata
import os

# Load spaCy visualizer
from spacy import displacy

# Import pandas DataFrame packages
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

# Import graphing package
import plotly.graph_objects as go
import plotly.express as px
```

```python
os.getcwd()
```

```python
texts = []
file_names = []

directory_path = '/Pasteur'
for _file_name in os.listdir(directory_path):
    if _file_name.endswith('.txt'):
        file_path = os.path.join(directory_path, _file_name)
        with open(file_path, 'r', encoding='utf-8') as file:
            texts.append(file.read())
            file_names.append(_file_name)
```

```python
d = {'Filename':file_names,'Text':texts}
letter_df = pd.DataFrame(d)
letter_df.head(12)
```

```python
letter_df['Text'] = letter_df['Text'].str.replace('\s+', ' ', regex=True).str.strip()
letter_df.head(12)
```

```python
metadata_df = pd.read_csv('/Metadata/Metadata_pasteur.csv')
metadata_df.head()
```

```python
letter_df['Filename'] = letter_df['Filename'].str.replace('.txt', '', regex=True)
```

```python
metadata_df.rename(columns={"LetterID": "Filename"}, inplace=True)
```

```python
metadata_df.head(12)
```

```python
final_letter_df = metadata_df.merge(letter_df,on='Filename')
```

```python
final_letter_df.head()
```

```python
nlp = spacy.load('fr_core_news_sm')
```

```python
print(nlp.pipe_names)
```

```python
phrase = "Ceci est une phrase exemple"

# Call the nlp model on the sentence
doc = nlp(phrase)
# Loop through each token in doc object
for token in doc:
    # Print text and part of speech for each
    print(token.text, token.pos_)
```

```python
def process_text(text):
    return nlp(text)
```

```python
final_letter_df['Doc'] = final_letter_df['Text'].apply(process_text)
```

```python
# Define a function to retrieve tokens from a doc object
def get_token(doc):
    return [(token.text) for token in doc]
```

```python
# Run the token retrieval function on the doc objects in the dataframe
final_letter_df['Tokens'] = final_letter_df['Doc'].apply(get_token)
final_letter_df.head()
```

```python
tokens = final_letter_df[['Text', 'Tokens']].copy()
tokens.head()
```

```python
# Define a function to retrieve lemmas from a doc object
def get_lemma(doc):
    return [(token.lemma_) for token in doc]

# Run the lemma retrieval function on the doc objects in the dataframe
final_letter_df['Lemmas'] = final_letter_df['Doc'].apply(get_lemma)
```
